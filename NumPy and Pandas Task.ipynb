# NumPy

```python
import numpy as np
```

#### Q1. Write a NumPy program to create a new shape to an array without changing its data.

```python
Arr1 = np.array([1, 21, 32, 43, 54, 65, 5, 44, 10, 11, 12, 13])
new_array = Arr1.reshape((2,3,-1))
new_array
```

#### Q2. Write a NumPy program to change the data type of an array.

```python
Arr2 = np.array([[24, 4, 6, 3], [61, 8, 10, 91]], np.int64)
Arr2 = Arr2.astype(dtype = str)
Arr2
```

#### Q3. Write a NumPy program to create an array of 3,5 shape and swap column 2 with column 3.

```python
Arr3 = np.arange(15).reshape(3,5)
print(Arr3)
Arr3[:,(2,3)] = Arr3[:,(3,2)]
print("\n",Arr3)
```

#### Q4. Write a NumPy program to create one-dimensional array of four, five and six digit numbers.

```python
four_digit_Arr = np.arange(4)
five_digit_Arr = np.arange(5)
six_digit_Arr = np.arange(6)
print(four_digit_Arr)
print(five_digit_Arr)
print(six_digit_Arr)
```

#### Q5. Write a NumPy program to find the missing data in a given array. 

```python
# if we want to find some values we need to find their indices

Arr4 = np.array([[3, 2, np.nan, np.nan],[10, np.nan, 10, np.nan],[5, np.nan, 1, np.nan],[25, 3, 1, np.nan]])
missing_values = np.where(np.isnan(Arr4))
missing_values = (np.dstack(missing_values)).reshape(-1,2)
missing_values
# list(Arr4).ind
```

#### Q6. Write a NumPy program to add a new row to an empty NumPy array.

```python
empty_array = np.empty((4,3), float)
print(empty_array)
appended_array = np.append(empty_array, np.zeros((1,3)), axis = 0)
print("\n",appended_array)
```

#### Q7. Write a NumPy program to convert a Python dictionary to a NumPy ndarray. 

```python
d = {"a" : 5, "b" : 4, "r" : 5, "t" : 2}
dict_to_arr = np.array(list(d.items()))
# dict_to_arr
dict_to_arr
```

#### Q8. Write a NumPy program to check whether the dimensions of two given arrays are same or not.

```python
array1 = np.ones((5,4,2))
array2 = np.ones((2,3))
if array1.ndim == array2.ndim:
    print("array1 and array2 have the same dimensions")
else:
    print("array1 and array2 don't have the same dimensions")
```

#### Q8. Write a Pandas program to join the two given Data Frames along rows and assign all data.
