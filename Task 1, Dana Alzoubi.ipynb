# Q1

```python
#without nested loop
li = list(range(1,6)) #1,2,3,4,5

for i in li:
    if i == li[(len(li)-1)]:
        print("*"*(i-3),end="")
        print("  ",end="")
        print("*")
    else:
        print("*"*i)
for i in li[-2::-1]:
    print("*"*i)
    
#Jordan Flag Triangle <3
```

```python
#using nested loop

li = list(range(1,6)) #1,2,3,4,5
for i in li:
    for j in range(i):
        print("*",end="")
    print()
for i in li[-2::-1]: #to exclude the last item "5" we start from -2
    for j in range(i):
        print("*",end="")
    print()
```

```python
#the summation and the average of a number's digits
#I solve the previous question as we have to calculate the sum of one number's digits.. so i kept it here as a new one.

n = int(input())
sum = 0
loops = 0
while n>0:
    digit = n%10
    n //= 10
    sum += digit
    loops += 1
print("Sum = "+str(sum)+"\nAvg = "+str(sum/loops))
```

# Q4

```python
password = input()

more_than_min = len(password) >= 6
less_than_max = len(password) <= 16

has_digit = False
has_letter = False
has_special_char = False

for char in password:
    if char.isdigit():
        has_digit = True
    elif char.isalpha():
        has_letter = True
    elif char in ["$","#","@"]:
        has_special_char = True
    if has_digit and has_letter and has_special_char:
        break

if more_than_min and less_than_max and has_digit and has_letter and has_special_char:
    print("Valid Password: "+str(password))
else:
    print("Non Valid Password: "+str(password))
    

# let it until i study the whole material
```

```python
String = input()
count = 0
for i in String:
    if i.isdigit() or i.isalpha():
        count += 1
print(count)
```

```python
#using the ASCII code
String = input()
count = 0
for i in String:
    if (ord(i) >= 65 and ord(i) <= 90) or (ord(i) >= 97 and ord(i) <= 122) or (ord(i) >= 48 and ord(i) <= 57):
        count += 1
print(count)
# print(ord("A"),ord("Z"),ord("a"),ord("z"),ord("0"),ord("1"),ord("9"))
```
